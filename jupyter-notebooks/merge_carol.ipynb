```python
from os import listdir
import pandas as pd
```

```python
path = '../data/carol'

dirs = [path + '/' + dir for dir in listdir(path) if dir.startswith('cases')]    # list of directories
files = [dir + '/' + listdir(dir)[1] for dir in dirs]    # list of paths to .csv files in all directories

# Read first .csv
data = pd.read_csv(files[0])
files = files[1:]

# Append remaining .csv files to first file
for f in files:
    data = pd.concat([data, pd.read_csv(f)], axis=0)

# Restrict to accidents
data = data.loc[data['EventType']=='ACC']

# Write merged file to .csv
data.to_csv(path+'carol_merged.csv')
```
