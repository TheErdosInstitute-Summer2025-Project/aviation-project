# Read & Prepare 'Findings' Dataset for Joining

```python
import pandas as pd
```

```python
df = pd.read_csv('../data/ntsb/ntsb_injuries.csv',usecols=['ev_id',
                                                             'Aircraft_Key',
                                                             'inj_person_category',
                                                             'injury_level',
                                                             'inj_person_count'])


```

```python
df = df[~df['inj_person_count'].isnull()]
```

```python
## Set new id for each case
df['event_key'] = df['ev_id'].astype(str) + '_' + df['Aircraft_Key'].astype(str)

```

```python
print(list(set(df['injury_level'].values)))
injury_levels = ['NONE', 'MINR', 'FATL', 'SERS']
injured_levels = ['MINR', 'FATL', 'SERS']
```

```python
total_person_df = (
    df[df['injury_level'].isin(injury_levels)]
    .groupby('event_key')['inj_person_count']
    .sum()
    .reset_index()
    .rename(columns={'inj_person_count': 'total_person_count'})
)

injured_person_df = (
    df[df['injury_level'].isin(injured_levels)]
    .groupby('event_key')['inj_person_count']
    .sum()
    .reset_index()
    .rename(columns={'inj_person_count': 'injured_person_count'})
)

FATL_person_df = (
    df[df['injury_level'] == 'FATL']
    .groupby('event_key')['inj_person_count']
    .sum()
    .reset_index()
    .rename(columns={'inj_person_count': 'Fatal_count'})
)

MINR_person_df = (
    df[df['injury_level'] == 'MINR']
    .groupby('event_key')['inj_person_count']
    .sum()
    .reset_index()
    .rename(columns={'inj_person_count': 'Minor_count'})
)

SERS_person_df = (
    df[df['injury_level'] == 'SERS']
    .groupby('event_key')['inj_person_count']
    .sum()
    .reset_index()
    .rename(columns={'inj_person_count': 'Serious_count'})
)
```

```python
injury_count_df = total_person_df.merge(injured_person_df, on='event_key', how='left')
injury_count_df = injury_count_df.merge(MINR_person_df, on='event_key', how = 'left')
injury_count_df = injury_count_df.merge(SERS_person_df, on='event_key', how = 'left')
injury_count_df = injury_count_df.merge(FATL_person_df, on='event_key', how = 'left')
```

```python
injury_count_df = injury_count_df.fillna(0)
injury_count_df['ev_id'] = injury_count_df['event_key'].str.split('_').str[0]
injury_count_df['Aircraft_key'] = injury_count_df['event_key'].str.split('_').str[1]
injury_count_df.head(50)
```

```python
injury_count_df.to_csv('../data/ntsb/cleaned/injuries_data.csv')
```
