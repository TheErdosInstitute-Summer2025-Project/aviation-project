```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../data/ntsb/ntsb_aircraft.csv',
                   usecols=['far_part','acft_make','acft_category','homebuilt',
                            'total_seats','num_eng','fixed_retractable','date_last_insp',
                            'type_fly','second_pilot','rwy_len','rwy_width',
                            'fuel_on_board','unmanned'])
```

```python
planes = data.loc[data['acft_category']=="AIR"]

for col in planes.columns:
    print(col)
    print(planes[col].describe())
    print()
    print()
```

```python
small_planes = planes.loc[planes['total_seats']<20]
plt.boxplot(small_planes['total_seats'])
plt.show()
```
