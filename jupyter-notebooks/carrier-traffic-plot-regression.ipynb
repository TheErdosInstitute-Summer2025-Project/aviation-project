```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
flights = pd.read_csv('us-carrier-traffic.csv')

flights["time"] = pd.to_datetime(flights.Period)

# extract mm from 'yyyy-mm-dd' and converts to int
month = flights.Period.apply(lambda x: int(x[5:7]))
flights["month"] = month

# transform month to be sinusoidal with period 12, min @ x = 2
flights["month_feat"] = month.apply(lambda x: -math.cos((x-2)* 2*math.pi / 12))
```

```python
plt.plot(flights.time, flights.Total)
plt.xlabel("Year", fontsize=14)
plt.ylabel("Flights / month (in tens of millions)", fontsize=14)
plt.title("US domestic commercial flights by month", fontsize=14)

plt.show()
```
